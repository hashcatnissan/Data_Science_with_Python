# **Querying A Series in Pandas**

```python
s1 = pd.Series([x for x in range(1,11)])
```

```python
s1
```

```python
s1.iloc[0]
```

```python
s1.iloc[8]
```

```python
s1.iat[0]
```

```python
s1.iat[8]
```

```python
s1[5:9]
```

```python
s1[-4:-1]
```

```python
s1.where(s1%2==0)
```

```python
s1.where(s1%2==0,'Odd Number')
```

```python
s1.where(s1%2==0,s1**2)
```

```python
s1.where(s1%2==0, inplace=True)
```

```python
s1.dropna()
```

```python
s1.fillna('Filled Value')
```
