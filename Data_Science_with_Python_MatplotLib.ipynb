# **MatPlotLib for 3D Visualization**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/My Project/Data Science with Python/HousePrices.csv',parse_dates=True)
```

```python
df.head(5)
```

```python
df['Y-S'] = df.YrSold	 - df.SalePrice
df['100MA'] = df['LotArea'].rolling(100).mean()
```

```python
#Plotting 3D Graph
ax = plt.axes(projection='3d')
ax.scatter(df.index,df['Y-S'],df['100MA'])
ax.set_xlabel('Index')
ax.set_ylabel('Y-S')
ax.set_zlabel('100MA')
plt.show
```

```python
#Plotting 3D Graph
ax = plt.axes(projection='3d')
ax.scatter(df.index,df['Y-S'],df['MoSold'])
ax.set_xlabel('Index')
ax.set_ylabel('Y-S')
ax.set_zlabel('MoSold')
plt.show
```
